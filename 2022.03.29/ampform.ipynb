## Build amplitude model


```python
import qrules

reaction = qrules.generate_transitions(
    initial_state=("J/psi(1S)", [-1, +1]),
    final_state=["K0", "Sigma+", "p~"],
    allowed_intermediate_particles=["Sigma(1660)"],
    allowed_interaction_types="strong",
)
```

```python
import ampform
import sympy as sp
from ampform.dynamics.builder import create_relativistic_breit_wigner_with_ff

builder = ampform.get_builder(reaction)
for particle in reaction.get_intermediate_particles():
    builder.set_dynamics(
        particle.name, create_relativistic_breit_wigner_with_ff
    )
model = builder.formulate()
model.intensity
```

```python
from IPython.display import Math

symbol, expr = list(model.components.items())[0]
latex = sp.latex(expr.doit() * 12 / sp.sqrt(6) / -1)
Math(latex)
```
