```python
import graphviz
import sympy as sp
from IPython.display import display
```

```python
g, n = sp.symbols("g_a n_a")
s, m0, w0 = sp.symbols("s m0 Gamma0")
expr = g * n / (m0**2 - s - w0 * m0 * sp.I)
dot = sp.dotprint(expr, bgcolor="none", size=3)
display(expr)
graphviz.Source(dot)
```
